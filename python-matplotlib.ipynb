# Python-Matplotlib
- Matplotlib 库太大，画图通常只需其中的核心模块 matplotlib.pyplot，并给其一个别名 plt，即 import matplotlib.pyplot as plt；

- 为了使图形在展示时能够很好的嵌入到 Jupyter 的 Out[] 中，需要使用 %matplotlib inline。

## 绘图基础
### 绘制图像

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 绘制图像
Fig1 = plt.figure()
x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]
plt.plot(x, y)
```

这里绘制虽然没问题，但遗憾的是图形太浑浊，下面的示例展示高清的 svg 矢量图。

```python
# 展示高清图代码
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('svg')
# 绘制图像
Fig1 = plt.figure()
x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]
plt.plot(x, y)
```

### 保存图像
保存图像用 savefig() 方法，其需要一个 r 字符串：r'绝对路径\图像名.后缀'。
- 绝对路径：如果要保存到桌面，绝对路径即：C:\Users\用户名\Desktop;
- 后缀：可保存图片的格式很多，包括 eps、jpg、pdf、png、svg（推荐）等。

保存为 svg 格式后，可直接拖至 Word 或 Visio 中，即可显式矢量高清图。

### 两种画图方式
Matplotlib 有两种画图方式：Matlab 方式和面向对象方式。这两种方式都可以完成同一个目的，也可以相互转化。

```python
# 准备数据
x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]

# Matlab方式
Fig1 = plt.figure()
plt.plot(x, y)

# 面向对象方式
Fig2 = plt.figure() # 图窗
ax2 = plt.axes() # 坐标轴
ax2.plot(x, y)
```
