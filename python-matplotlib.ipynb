## 绘图基础
### 绘制图像

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 绘制图像
Fig1 = plt.figure()
x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]
plt.plot(x, y)
```

### 两种画图方式
Matplotlib 有两种画图方式：Matlab 方式和面向对象方式。这两种方式都可以完成同一个目的，也可以相互转化。

```python
# 准备数据
x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]

# Matlab方式
Fig1 = plt.figure()
plt.plot(x, y)

# 面向对象方式
Fig2 = plt.figure() # 图窗
ax2 = plt.axes() # 坐标轴
ax2.plot(x, y)
```

## 多图形的绘制
在 Jupyter 的某个代码块中使用 Fig1 = plt.figure() 创建图窗后，其范围仅仅在此代码块内，跳出代码块外的其他画图命令将于 Fig1 无关。因此，画一幅图，请在一个代码块内完成。

### 绘制多线条
在同一个图窗内绘制多线条，按两种画图方式分开来演示。

```python
# 准备数据
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [0, 0, 0, 0, 0]
y3 = [-1, -2, -3, -4, -5]

# Matlab方式
Fig1 = plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

# 面向对象方式
Fig2 = plt.figure()
ax2 = plt.axes()
ax2.plot(x, y1)
ax2.plot(x, y2)
ax2.plot(x, y3)
```

### 绘制多个子图


```python
# 准备数据
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [0, 0, 0, 0, 0]
y3 = [-1, -2, -3, -4, -5]

# Matlab方式
Fig1 = plt.figure()
plt.subplot(3, 1, 1), plt.plot(x, y1)
plt.subplot(3, 1, 2), plt.plot(x, y2)
plt.subplot(3, 1, 3), plt.plot(x, y3)
```

```python
# 面向对象方式
Fig2, ax2 = plt.subplots(3) # 一行替代之前的两行代码
ax2[0].plot(x, y1)
ax2[1].plot(x, y2)
ax2[2].plot(x, y3)
```

## 图表类型
常用图表类型：二维图，网格图，统计图。

- 二维图只需要两个向量即可绘图。包括线型图（plot），散点图（scatter），条形图（bar），杆图（stem），阶梯图（step），误差图（fill_between），堆叠图（stackplot）等，具体可见[Matplotlib官方文档](https://matplotlib.org/stable/plot_types/index)。

- 网格图只需要一个矩阵即可绘图。以下网格图都有一定的使用价值：图像展示（imshow），等高线（contour），填充等高线（contourf）等。

- 统计图一般做数据分析时使用。包括直方图（hist），箱型图（boxplot），二维直方图（hist2d），饼图（pie）等。

除了这些之外，还有更多作者提前画好的花哨的靓图，详见[花哨的靓图模板](https://matplotlib.org/stable/gallery/index.html)。最后，作者还温馨地向小白的我们提供了从 0 开始到大神的完整教程，详见[从0开始到画图大神](https://matplotlib.org/stable/tutorials/index.html)。不过作为一名技术工程师，不要在画图上花费太多时间。

### 二维图
仅演示 plot 线型图函数，只因其可以替代其他所有二维图。

#### 设置颜色，风格，粗细，标记
- 颜色：plot() 含 color 关键字，可以设置线条的颜色；
- 风格：plot() 含 linestyle 关键字，可以设置线条的风格；
- 粗细：plot() 含 linewidth 关键字，可以设置线条的粗细，一般 0.5-3 之间为宜；
- 标记：plot() 含 marker 关键字，可以设置线条的标记；
- 标记大小：plot() 含 markersize 关键字，可以设置标记的大小。 

```python
# 准备数据
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 3, 4, 5, 6]
y3 = [3, 4, 5, 6, 7]
y4 = [4, 5, 6, 7, 8]

# 设置颜色，风格，粗细，标记，标记大小
Fig2, ax2 = plt.subplots()
ax2.plot(x, y1, color='#7CB5EC', linestyle='-', linewidth=0.5, marker='.', markersize='6')
ax2.plot(x, y2, color='#F7A35C', linestyle='--', linewidth=1, marker='o', markersize='6')
ax2.plot(x, y3, color='#A2A2D0', linestyle='-.', linewidth=1.5, marker='^', markersize='6')
ax2.plot(x, y4, color='#111111', linestyle=' ', linewidth=2, marker='D', markersize='6')
```

### 网格图
这里仅演示 imshow() 函数，只因另外两个在深度学习中几乎用不到。

```python
import numpy as np
x = np.linspace(0, 10, 1000) # 这是一个包含0-10线性增加的1000个元素的向量 
l = np.sin(x) * np.cos(x).reshape(-1, 1)
Fig1 = plt.figure()
plt.imshow(l)
plt.colorbar()
```

注意，上面的 np.cos(x) 是形状为 (1000,) 的向量，两个向量乘积仍是向量，因此需要重塑一下，利用广播机制生成矩阵。重塑后 np.sin(x) 的形状为 (1000,)，np.cos(x).reshape(-1, 1) 的形状为 (1000, 1)，广播机制的过程如下：
1. np.sin(x) 首先被广播为 (1, 1000) 的行矩阵，与形状为 (1000, 1) 的列矩阵相乘；
2. np.sin(x) 和 np.cos(x).reshape(-1, 1) 都被广播为形状为 (1000, 1000) 的矩阵，再逐元素相乘得到结果。

### 统计图
统计图仅演示 hist 函数，只因其他函数主要出现在数据分析领域。hist 是统计学的函数，是为了看清某分布的均值和标准差等重要统计特征。

可通过函数形参关键字调节区间个数，透明度，图表类型（是否浑然一体），边缘颜色等。

```python
data1 = np.random.randn(1000)
data2 = np.random.normal(3, 1, 1000)
Fig1 = plt.figure()
plt.hist(data1, bins=30, alpha=0.8, histtype='bar', edgecolor='#FFFFFF')
plt.hist(data2, bins=30, alpha=0.5, histtype='stepfilled', color='#7CB5EC')
```

## 图窗属性
### 坐标轴上下限
现设定其坐标轴上下限，有两种方法：lim 法和 axis 法。此外可以用 plt.axis('equal') 使 x 轴和 y 轴的比例为 1:1。

```python
x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]

# lim法
Fig1 = plt.figure()
plt.plot(x, y)
plt.xlim(1, 5)
plt.ylim(1, 125)

# axis法
Fig2 = plt.figure()
plt.plot(x, y)
plt.axis([1, 5, 1, 125])
# plt.axis('equal')
```

### 标题与轴名称
在这里，Matlab 方式与面向对象方式将**最后一次出现区别**。

```python
x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]

# Matlab方式
Fig1 = plt.figure()
plt.plot(x, y)
plt.title('This is the title.')
plt.xlabel('This is xlabel.')
plt.ylabel('This is ylabel.')

# 面向对象方式
Fig2, ax2 = plt.subplots()
ax2.plot(x, y)
ax2.set_title('This is the title.')
ax2.set_xlabel('This is xlabel.')
ax2.set_ylabel('This is ylabel.')
```

### 图例和网格
一般图例会出现在二维图与统计图中，网格图则用的是颜色条。

主要使用 legend 函数，该函数有三个常用的关键字参数：loc，frameon，ncol
- loc 用于表示图例位置，在 upper，center，lower 中选一个，在left，center，right 中选一个，也可以 loc='best' 具体用法见下例；
- frameon 用于表示图例边框，去边框是 frameon=False；
- ncol 用于表示图例的列数，默认是 1 列，也可以通过 ncol=2 调为 2 列。

此外，可以用 grid() 加上网格。

```python
# 准备数据
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [0, 0, 0, 0, 0]
y3 = [-1, -2, -3, -4, -5]

Fig1 = plt.figure()
plt.plot(x, y1, label='y=x')
plt.plot(x, y2, label='y=0')
plt.plot(x, y3, label='y=-x')
plt.legend(loc='upper left', frameon=True, ncol=1)
plt.grid()
# plt.legend(['y=x', 'y=0', 'y=-x'])
```
